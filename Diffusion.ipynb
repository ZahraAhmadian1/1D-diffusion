```python
import numpy as np
import matplotlib.pyplot as plt

#This code is to solve 1D diffusion equation: du/dt= D d2u/dt2

#Initial constants
L=1
T=0.1
Nx=100
dt=0.0003
Nt=int(T/dt)
dx=L/Nx
D=0.1
Cons=(dt*D)/(dx*dx)
#stable solution only for Cons < 0.5
if Cons > 0.5:
    raise ValueError("Change dt or dx")
    
#boundary conditions u(0,t)=u(L,t)=0    
#initial condition
x=np.linspace(0,L,Nx)
u=np.sin(np.pi*x/L)
u_next=np.zeros(Nx)

#u_next is the updated u in the next time step.
for i in range(0,Nt):
    #boundary condition
    u_next[0]=0
    u_next[-1]=0
    #loop over x
    for j in range(1,Nx-1):
        u_next[j]=Cons*(u[j+1]-2*u[j]+u[j-1])+u[j]
    #end of loop
    u=u_next.copy()
    
plt.plot(x,u,'r')
plt.xlabel('x')
plt.ylabel('u(x,t)')

    
```
